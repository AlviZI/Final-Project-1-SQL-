```python
import csv
from faker import Faker
from faker_vehicle import VehicleProvider
from tabulate import tabulate

fake = Faker()
fake.add_provider(VehicleProvider)

headers = ['product_id', 'brand', 'model', 'body_type', 'year', 'price']
data = []

for i in range(49):
    row = [fake.random_int(min=1, max=14), fake.vehicle_make(), fake.vehicle_model(), fake.vehicle_category(), fake.vehicle_year(), fake.random_int(min=1000, max=100000)]
    data.append(row)

with open('/content/drive/MyDrive/Pacmann SQL/Final Project/car_product.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(headers)
    writer.writerows(data)

print(tabulate(data, headers=headers, tablefmt='grid'))
```

# Dummy Data City

```python
import csv
from faker import Faker
from tabulate import tabulate


fake = Faker('id_ID')

headers = ['kota_id', 'nama_kota', 'latitude', 'longitude']
data = []

for i in range(49):
    row = [fake.random_int(min=1, max=49), fake.city(), fake.latitude(), fake.longitude()]
    data.append(row)

with open('/content/drive/MyDrive/Pacmann SQL/Final Project/city.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(headers)
    writer.writerows(data)

print(tabulate(data, headers=headers, tablefmt='grid'))
```
